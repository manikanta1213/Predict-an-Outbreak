```python
import matplotlib.pyplot as plt

case='Confirmed'
def timeCompare(time,*argv):
    Coun1=argv[0]
    Coun2=argv[1]
    f,ax=plt.subplots(figsize=(16,5))
    labels=argv  
    country=df.loc[(df['COUNTRY/AREA/TERRITORY']==Coun1)]
    plt.plot(country['DATE'],country['INF_A'],linewidth=2)
    plt.xticks([])
    plt.legend(labels)
    ax.set(title=' Evolution of actual cases',ylabel='Number of cases' )

    country2=df.loc[df['COUNTRY/AREA/TERRITORY']==Coun2]
    #country2['Date']=country2['Date']-datetime.timedelta(days=time)
    plt.plot(country2['DATE'],country2['INF_A'],linewidth=2)
    #plt.xticks([])
    plt.legend(labels)
    ax.set(title=' Evolution of cases in %d days difference '%time ,ylabel='Number of %s cases'%case )
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
def fit_model(df,future):
    m = Prophet(daily_seasonality=True)
    m.fit(df)
    forecast = m.predict(future)
    return m, forecast, future

def fb_plots(m, fore):
    return plot_plotly(m, forecast)

def fb_subplots(m, fore):
    return m.plot(forecast), m.plot_components(forecast)
```

```python
data = pd.DataFrame()
data['ds'] = test['DATE']
data['y'] = test['INF_A']
train_data = data[:50]
test_data = data[50:]
future = pd.DataFrame()
future['ds'] = test_data['ds']
```

```python
model, forecast, future =fit_model(train_data, future)
```

```python
fb_plots(model, forecast)
```

```python
fb_subplots(model, forecast)
```

```python
forecast['yhat'], test_data['y']
```

```python
print("Root Mean Squared Error for Prophet Model: ",np.sqrt(mean_squared_error(test_data["y"],forecast['yhat'])))
```
